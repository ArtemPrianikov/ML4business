```python
data = pd.read_csv('my_PJME_MW.csv', index_col = [0],parse_dates=[0])
data.head()
```

Функция для расчёта скользящей средней

```python
def exponential_smoothing(series, alpha):
    result = [series[0]] # first value is same as series
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
```

из внутридневных данных создадим дневные с усреднением:

```python
d_data=data.resample('D').mean()
```

Рассчитаем экспоненциальную скользящую среднюю

```python
data_exp_1 = exponential_smoothing(d_data['PJME_MW']  , 0.05)
```

```python
plt.figure(figsize=(20, 9))
plt.plot(d_data.index[:500], d_data['PJME_MW'].values[:500], '.b',label='исходный ряд')

plt.plot(d_data.index[:500],d_data['PJME_MW'].values[:500] - data_exp_1[:500] , color = 'green', label="без сезона по экспоненциальному сглаживанию");

plt.plot(d_data.index[:500],data_exp_1[:500] , color = 'red', label="экспоненциальное сглаживание сезон");

plt.xlabel('time')
plt.ylabel('MW')
plt.legend()
plt.title('PJME_MW')
```

Ряд стал стационарным:

```python
df= pd.DataFrame(data_exp_1)

df['t'] = 0
df.iloc[1:,1]=df.iloc[:-1,0].values
df['t1']=0
df.iloc[2:,2]=df.iloc[:-2,0].values
df['t2']=0
df.iloc[3:,3]=df.iloc[:-3,0].values
df['t3'] = 0
df.iloc[4:,4]=df.iloc[:-4,0].values
df['t4']=0
df.iloc[5:,5]=df.iloc[:-5,0].values
df['t5']=0
df.iloc[6:,6]=df.iloc[:-6,0].values
df.fillna(0)

df.head(10)
```

```python
# добавили экспоненциальное сглаживание с тремя разными весами
df['exp_0_01'] = exponential_smoothing(df[0], 0.01)
df['exp_0_03'] = exponential_smoothing(df[0], 0.03)
df['exp_0_08'] = exponential_smoothing(df[0], 0.08)
```

Назначим индекс из дат (из подневного датафрейма)

```python
df.index = d_data.index
```

Создаю столбец с разницей, по-другому у меня почему-то не получилось, пришлось немного костыльно:

```python
ser1 = df.iloc[:-1,0]
ser2 = df.iloc[1:,0]
ser3 = [x[1]-ser2[x[0]] for x in enumerate(ser1)]
df['diff1'] = 0
df['diff1'].iloc[1:] = ser3
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score

scl =StandardScaler()
scly =StandardScaler()

def split_data_b(data, split_date ):
    return data.loc[data.index.get_level_values('Date') <= split_date].copy(), \
           data.loc[data.index.get_level_values('Date') >  split_date].copy()

train_b, test_b = split_data_b( df, '10-10-2014')

X_train_b = train_b.iloc[:-1,:]
y_train_b = train_b[df.columns[0]].values[1:]
X_test_b = test_b.iloc[:-1,:]
y_test_b = test_b[df.columns[0]].values[1:]

```

```python
plt.figure( figsize=( 15, 7 ))
plt.xlabel('Время')
plt.ylabel('мВт')
plt.plot(train_b.index.get_level_values('Date'),train_b[df.columns[0]], label='train data' )
plt.plot(test_b.index.get_level_values('Date'),test_b[df.columns[0]], label='test data')
plt.title( 'Тестовые и тренировочные данные' )
plt.legend()
plt.show()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

Обучим модель и посчитаем ошибку

```python
%%time
model_gb = GradientBoostingRegressor(max_depth=15, random_state=0, n_estimators=100)
model_gb.fit( X_train_b, y_train_b ) 
X_test_pred_gb = model_gb.predict(X_test_b)
```

```python
plt.figure( figsize=( 20, 4 ))

plt.xlabel('Время')
plt.ylabel('Потребление мВт')
plt.plot(test_b.index[1:],X_test_pred_gb, label='predict GB data' )
plt.plot(test_b.index[1:],y_test_b, label='test data')
plt.plot(test_b.index, test_b[[0]], label='исходный ряд')

plt.legend()
plt.show()

er_g = mean_absolute_percentage_error(y_true=y_test_b, y_pred=X_test_pred_gb)

```

## Задание 3

Сделайте вывод о том, как изменилась ошибка модели по значениям h.

Видно, что с увеличением "окна" предсказания, растёт ошибка. Интересно посмотреть, как соотносятся p и h, как их различные сочетания влияют на ошибку? Например, с увеличением кол-ва учитываемой истории, должна уменьшаться ошибка при прогнозировании на большем окне.

Пока попробуем сделать предсказания на разных горизонтах (h) и построить наглядный график как меняется ошибка:

```python
%%time
h_list = []
for h in range (1,10):
    X_train_b = train_b.iloc[:-h,:]
    y_train_b = train_b[df.columns[0]].values[h:]
    X_test_b = test_b.iloc[:-h,:]
    y_test_b = test_b[df.columns[0]].values[h:]

    model_gb.fit( X_train_b, y_train_b ) 
    X_test_pred_gb = model_gb.predict(X_test_b)
    
    er_g_ = mean_absolute_percentage_error(y_true=y_test_b, y_pred=X_test_pred_gb)
    
    item = {'h_value':h, 'MAPE':er_g_}
    h_list.append(item)
    print(item)
```

```python
h_df = pd.DataFrame(h_list)
```

```python
plt.figure( figsize=(8, 4 ))

plt.xlabel('h_value')
plt.ylabel('MAPE')
plt.plot(h_df['h_value'],h_df['MAPE'], label='predict GB data' )

plt.legend()
plt.show()
```

Хорошие новости! Ошибка растёт линейно, не экспоненциально!

```python
%%time
columns = [2,3,4,5,6]
p_list = []
h = 1
for p in range(1,6):
    df_temp = df.copy()
    #df_temp = df.iloc[:,1:columns]
    df_temp = df_temp.drop(df_temp.columns[columns[:6-p]], axis = 1)
    
    train_b, test_b = split_data_b(df_temp, '10-10-2014')
    
    X_train_b = train_b.iloc[:-h,:]
    y_train_b = train_b[df.columns[0]].values[h:]
    X_test_b = test_b.iloc[:-h,:]
    y_test_b = test_b[df.columns[0]].values[h:]

    model_gb.fit( X_train_b, y_train_b ) 
    X_test_pred_gb = model_gb.predict(X_test_b)
    
    er_g_ = mean_absolute_percentage_error(y_true=y_test_b, y_pred=X_test_pred_gb)
    
    item = {'p':p, 'MAPE':er_g_}
    p_list.append(item)
    print(item)
```

```python
p_df = pd.DataFrame(p_list)

plt.figure( figsize=(8, 4 ))

plt.xlabel('p_value')
plt.ylabel('MAPE')
plt.plot(p_df['p'],p_df['MAPE'], label='predict GB data' )

plt.legend()
plt.show()
```

Пока странный для меня результат, я думал, что с увеличением кол-ва данных (увеличения окна подсматривания назад) ошибка должна уменьшаться, а получается пока наоборот, но, наверно, тут нужны гораздо более подробные исследования
